```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
plt.style.use("classic")
```

$\large |x|^{\frac{2}{3}}+\frac{1}{3}e(\pi - x^2)^{\frac{1}{2}}\sin(a\pi x)$

```python
def F(a):
    f = lambda x: abs(x) ** (2 / 3) + np.e / 3 * (np.pi - x ** 2) ** (1 / 2) * np.sin(
        a * np.pi * x
    )
    return f
```

```python
def update(a):
    global x
    line.set_ydata(F(a)(x))
    ax.set_title(
        r"$|x|^{\frac{2}{3}}+\frac{1}{3}e(\pi -x^2)^{\frac{1}{2}}\sin(a\pi x)$"
        + ", $a=%.1f$" % a,
        fontsize=30,
        pad=30,
    )
    plt.tight_layout()
    return line, ax
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
x = np.linspace(-np.sqrt(np.pi), np.sqrt(np.pi), 1000)
(line,) = ax.plot(x, F(1)(x), color="#ed1c24", lw=5)
ax.grid()
ax.set_ylim([-1.6, 2.5])
ax.set_title(
    r"$|x|^{\frac{2}{3}}+\frac{1}{3}e(\pi -x^2)^{\frac{1}{2}}\sin(a\pi x)$"
    + ", $a=0.1$",
    fontsize=30,
    pad=30,
)
plt.tight_layout()
anim = FuncAnimation(fig, update, frames=np.arange(0.1, 30, 0.1), interval=50)
#HTML(anim.to_html5_video())
```
